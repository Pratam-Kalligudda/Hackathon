```python
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
links = pd.read_csv('links.csv')
```

```python
links.rename(columns={'tmdbId': 'id'}, inplace=True)
```

```python
movies = movies.merge(links[['movieId', 'id']], on = 'id', how = 'left')
```
