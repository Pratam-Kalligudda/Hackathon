## Popularity Based Recommender System

```python
ratings_with_name = movies_2
```

```python
ratings_with_name.head()
```

```python
num_rating_df = ratings_with_name.groupby('title').count()['rating'].reset_index()
num_rating_df.rename(columns={'rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('title').mean()['rating'].reset_index()
avg_rating_df.rename(columns={'rating':'avg_rating'},inplace=True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df,on='title')
popular_df
```

```python
popular_df = popular_df[popular_df['num_ratings']>=250].sort_values('avg_rating',ascending=False).head(50)
```

```python
popular_df.head()
```

```python
movies_2.head()
```

```python
popular_df = popular_df.merge(movies_2,on='title').drop_duplicates('title')[['id','movieId','userId','title','rating','num_ratings','avg_rating']]
```

## Collaborative Filtering Based Recommender System

```python
x = ratings_with_name.groupby('userId').count()['rating'] > 50
padhe_likhe_users = x[x].index
```

```python
filtered_rating = ratings_with_name[ratings_with_name['userId'].isin(padhe_likhe_users)]
```

```python
y = filtered_rating.groupby('title').count()['rating']>=50
famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['title'].isin(famous_books)]
```

```python
pt = final_ratings.pivot_table(index='title',columns='userId',values='rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores.shape
```

```python
def recommend(movie_name):
    # index fetch
    index = np.where(pt.index==movie_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:5]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = movies_2[movies_2['title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('title')['title'].values))
        item.extend(list(temp_df.drop_duplicates('title')['movieId'].values))
        item.extend(list(temp_df.drop_duplicates('title')['id'].values))
        
        data.append(item)
    
    return data
```

```python
recommend('X-Men')
```
